## 3) Adding a third variable with hue

```python
# vídeo

# built-in dataset in seaborn
tips = sns.load_dataset("tips")
tips.head()

# scatterplot (with "hue")

hue_colors = {"Yes": "black", "No": "r"}
hue_colors = {"Yes": "#808080", "No": "#00FF00"}

sns.scatterplot(x="total_bill", y="tip", data=tips, 
                hue="smoker", hue_order=["Yes", "No"], palette=hue_colors)
plt.show()

sns.countplot(x="smoker", data=tips, hue="sex")
plt.show()
```

# 2) Visualizing Two Quantitative Variables

## 1) Introduction to relational plots and subplots

```python
# vídeo

# relationship plot (we'll use this from now on)

# col
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter",
           col="smoker")
plt.show()
```

```python
# row
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter",
           row="smoker")
plt.show()
```

```python
# col & row
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter",
           col="smoker", row="time")
plt.show()
```

```python
# col_wrap and col_order
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter",
           col="day", col_wrap=2, col_order=["Thur", "Fri", "Sat", "Sun"])
plt.show()
```

## 2) Customizing scatter plots

```python
# video

# Subgroups with point size
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter",
           size="size", hue="size")
plt.show()
```

```python
# Subgroups with point style
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter",
           style="smoker", hue="smoker")
plt.show()
```

```python
# Changing point transparency

# set alpha to be betwwen 0 and 1
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter",
           alpha=0.4)  # overlaping areas
plt.show()
```

## 3) Introduction to line plots

```python
# video

sns.relplot(x="model_year", y="mpg", data=mpg, kind="line",
            style="origin", hue="origin", 
            markers=True, dashes=False)  
plt.show()

# ---> 95% confident that the mean (default measure) is within this interval
```

```python
# Replacing confidence interval with standard deviation
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line",
            errorbar="sd")  # to set off --> errorbar=None
plt.show()
```

# 3) Visualizing a Categorical and a Quantitative Variable

## 1) Count plots and bar plots

```python
# video

# Categorical plots (count plots, bar plots)

# catplot()
category_order = ['no time at all', 'less than an hour a day', 'few hours a day', 'most of the day',]
sns.catplot(x="Internet usage", data=surveys, kind="count", order=category_order)
plt.xticks(rotation=45)
plt.show()
```

```python
# bar plots

sns.catplot(x="day", y="total_bill", data=tips, kind="bar", hue="day")
plt.show()

# Lines show 95% confidence intervals for the mean. To turn it off, errorbar=None
```

```python
# Separate into column subplots based on age category
sns.catplot(y="Internet usage", data=surveys, kind="count", col="Age")

# Show plot
plt.show()
```

## 2) Box plots

```python
# vídeo

sns.catplot(x="time", y="total_bill", data=tips, kind="box", 
            hue="time", 
            order=["Dinner", "Lunch"],
            whis=[5, 95]) # or 2.0 (2.0 * IQR)
plt.show()

print("By default, the whiskers extend to 1.5 * IQR (25th to 75th percentile)")
```

## 3)  Point plots

```python
# vídeo

# Point plot has categorical variable on x-axis
sns.catplot(x="smoker", y="total_bill", data=tips, kind="point")
plt.show()
```

```python
sns.catplot(x="smoker", y="total_bill", data=tips, kind="point", estimator=np.median,
           capsize=0.2)  #errorbar=None
plt.show()
```

## 2) Adding titles and labels: Part 1

```python
# vídeo

# Type: FacetGrid (relplot, catplot) --> Can create subplots
# vs 
# AxesSubplot (scatterplot, countplot, etc) --> Only creates a single plot

# FacetGrid
g = sns.catplot(x="Region", y="Birthrate", data=countries, kind="box")
print(type(g))

g.fig.suptitle("New title", y=1.03)
```

## 3) Adding titles and labels: Part 2

```python
# vídeo

# AxesSubplot
g = sns.boxplot(x="Region", y="Birthrate", data=countries)
print(type(g))

g.set_title("New title", y=1.03)
```

```python
# Titles for subplots
sns.set_style("whitegrid")
sns.set_palette("RdBu")

# FacetGrid
g = sns.catplot(x="Region", y="Birthrate", 
                data=countries[countries["Climate"].isin(['1', '2', '3'])], 
                kind="box", col="Climate")
print(type(g))

g.fig.suptitle("New general title", y=1.03)
g.set_titles("This is {col_name}")
g.set(xlabel="New X Label", ylabel="New Y Label")

plt.xticks(rotation=90)

plt.show()
```
