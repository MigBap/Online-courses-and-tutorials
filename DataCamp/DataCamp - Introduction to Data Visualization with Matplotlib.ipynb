## 2) Plotting time-series with different variables

```python
#vídeo

fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"])
ax.plot(climate_change.index, climate_change["relative_temp"])

ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm) / Relative temperature (Celsius)')
plt.title("Not the best scale!")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"], color="darkblue")
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color="darkblue")

# they share the same X axis, the Y axis are separate!
ax2 =ax.twinx()

ax2.plot(climate_change.index, climate_change["relative_temp"], color="r")
ax2.set_xlabel('Time')
ax2.set_ylabel('Relative temperature (Celsius)', color="r")
plt.title(" ax.twinx() --> same X axis, the Y axis are separate!")

# even the ticks can be colored!
ax.tick_params('y', colors='darkblue')
ax2.tick_params('y', colors='r')

plt.show()
```

```python
# create a function
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.tick_params('y', colors=color)

fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change["co2"], 'blue', 'Time', 'CO2 (ppm)')

ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change["relative_temp"], 'red', 'Time', 
                'Relative temperature (Celsius)')

plt.show()
```

## 2) Quantitative comparisons: histograms

```python
# load data
display(summer2016.head(3))

# create dataframes
mens_rowing = summer2016[(summer2016["Sport"] == "Rowing") & (summer2016["Sex"] == "M")]
mens_gymnastics = summer2016[(summer2016["Sport"] == "Gymnastics") & (summer2016["Sex"] == "M")]
```

```python
# vídeo

fig, ax = plt.subplots()

ax.bar("Rowing", mens_rowing["Height"].mean())
ax.bar("Gymnastics", mens_gymnastics["Height"].mean())
ax.set_ylabel("Height (cm)")
plt.show()
```

```python
# Introducing histograms

fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="Rowing", bins=5)
ax.hist(mens_gymnastics["Height"], label="Gymnastics", bins=5)

ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of obervations")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="Rowing", 
        bins=[150, 160, 170, 180, 190, 200, 210],
        histtype="step")


ax.hist(mens_gymnastics["Height"], label="Gymnastics", 
        bins=[150, 160, 170, 180, 190, 200, 210],
        histtype="step")

ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of obervations")
plt.title("YES, there are rowing men with < 170cm")
ax.legend()
plt.show()
```

## 3) Statistical plotting

```python
# vídeo

# Adding error bars to bar charts
fig, ax = plt.subplots()

ax.bar("Rowing",
       mens_rowing["Height"].mean(), 
       yerr=mens_rowing["Height"].std())

ax.bar("Gymnastics",
       mens_gymnastics["Height"].mean(), 
       yerr=mens_gymnastics["Height"].std())

ax.set_ylabel("Height (cm)")
plt.title("Error bars in a bar chart \n Height mean with std (additional vertical marker)")
plt.show()
```

```python
# Adding error bars to LINE charts
fig, ax = plt.subplots()

ax.errorbar(seattle_weather["DATE"],
            seattle_weather["MLY-TAVG-NORMAL"], 
            yerr=seattle_weather["MLY-TAVG-STDDEV"], label="Seattle")

ax.errorbar(austin_weather["DATE"],
            austin_weather["MLY-TAVG-NORMAL"], 
            yerr=austin_weather["MLY-TAVG-STDDEV"], label="Austin")

ax.set_ylabel("Temperature (Fahrenheit)")
ax.legend()
# plt.title("Error bars in a bar chart \n Height mean with std (additional vertical marker)")
plt.show()
```

```python
# Adding boxplots

fig, ax = plt.subplots()

ax.boxplot([mens_rowing["Height"], mens_gymnastics["Height"]])
ax.set_xticklabels(["Rowing", "Gymnastics"])
ax.set_ylabel("Height (cm)")
plt.title("Boxplot \n(invented by John Tukey, arguably the first data scientist)")
plt.show()
```

Guidelines for choosing plotting style

 - Avoid dark backgrounds
 - consider colorblind-friendly, if color is important
 - consider ink if someone is going to print
 - if printed in black and white, use the "grayscale" style

## 3) Automating figures from data

```python
# vídeo

sports = summer2016["Sport"].unique()
print(sports)

fig, ax = plt.subplots()
for sport in sports[:15]:
    sport_df = summer2016[summer2016["Sport"] == sport]
    ax.bar(sport, sport_df["Height"].mean(),
           yerr=sport_df["Height"].std())

ax.set_ylabel("Height (cm)")
ax.set_xticklabels(sports, rotation=90)
plt.show()
```
