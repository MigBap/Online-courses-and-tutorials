```python
#vídeo
#Pandas - Part II

# Index and Select Data
#Square brackets
#Advanced methods: loc , iloc


# ------->  Column Access []
print(brics["country"])   #aparece Name: country, dtype: object     # what type of object? 

print("")
print(type(brics["country"]))   #<class 'pandas.core.series.Series'> --->  panda Series
#in other words, pasting together a bunch of series you create a pd.DataFrame

print("")
print(brics[["country"]])  #já só aparece a informação, sem as labels       
print(type(brics[["country"]]))  #<class 'pandas.core.frame.DataFrame'>     
#1D labelled array até aqui

print("")
print(brics[["country", "capital"]])


#para selecionar linhas/observações
print("")
print(brics[1:4])

#Em resumo:
# Square brackets: 
#  1) limited functionality
#  2) idealmente, queremos algo como os 2D Numpy arrays: my_array[rows, columns]

# com Pandas --> usar o loc (label-based) e o iloc (integer position-based)

# --------> loc:
print("")
print(brics.loc["RU"])  #aparece-nos uma coluna, 1D   ---- <class 'pandas.core.series.Series'>
print("")
print(brics.loc[["RU"]]) # já aparece a "RU" e correspondentes valores associados ---- <class 'pandas.core.frame.DataFrame'>

print("")
print(brics.loc[["RU", "IN", "CH"]])


#Row and Column loc
print("")
print(brics.loc[["RU", "IN", "CH"], ["country", "area"]])  

print("")
print(brics.loc[:, ["country", "area"]])  #para escolher todas as observações das 2 variáveis


# Resumindo

#Square brackets:
#column access ----- brics[["country", "capital"]]
#row access ------- brics[1:5]   only trough slicing

#loc (label-based)
# row access ------- brics.loc[["RU", "IN", "SA"]]
# column access ------ brics.loc[ :, ["capital", "area"] ]                         # brics.loc[["capital", "area"]]  não dá
# row & column access ------ brics.loc[["RU", "SA"], ["area", "population"]]

print("")
#Row access iloc (integer position-based)

print("")
print(brics.loc[["RU"]]) #é igual a 
print(brics.iloc[[1]])

print("")
print(brics.loc[["RU", "CH", "SA"]])
print(brics.iloc[[1, 3, 4]])

print("")
print(brics.loc[["RU", "IN", "CH"], ["country", "area"]])  
print(brics.iloc[[1,2,3], [0,2]])

print("")
print(brics.loc[:, ["country", "area"]]) 
print(brics.iloc[:, [0,2]])
```

```python
#vídeo
#Loop Data Structures - Part I

# We saw how looping works with lists and strings. What about other data structures, such as dictionaries and numpr arrays?

world = {"afghanistan": 30.55, "albania":2.77, "algeria":39.21}

#how to print the key and the corresponding value?

for key, value in world.items() :  #tem que ter .items, senão não dá
    print(key + "--" + str(value))

#dictionaries are inherently unordered, por isso nunca aparece o output pela ordem em q as keys e values foram definidas
#os nomes key e value são tb arbitrários

#Com numpy arrays:
import numpy as np
np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
bmi = np_weight / np_height ** 2

for val in bmi:          #com 1D dá
    print(val)

#Com 2D numpy array:
meas = np.array([np_height, np_weight])

for val in meas:         #com 2D não dá, simplesmente faz print dos dois vetores
    print(val)

for val in np.nditer(meas):   #assim já dá, 10 printouts - 5 heights e depois os 5 weights
    print(val)

    
#Resumindo:

# 1) Dictionary:
       # for key, val in "dicionário_criado".items() :

# 2) Numpy array
       # for val in np.nditer(np_array_criado) :
```

```python
#vídeo
#Looping Data Structures - Part II

dicionário_brics = {
    "country":["Brazil", "Russia", "India", "China", "South Africa"], 
    "capital":["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
    "area":[8.516, 17.10, 3.286, 9.597, 1.221],
    "population":[200.4, 143.5, 1252, 1357, 52.98] }

import pandas as pd
BRICS = pd.DataFrame(dicionário_brics)
BRICS.index = ("BR", "RU", "IN", "CH", "SA")
BRICS

for val in BRICS:
    print(val)   # -------------->  imprime simplesmente os nomes da tabela/variáveis
print("")


for label, row in BRICS.iterrows() :  #---> temos que utilizar o .iterrows()
    print(label)    #faz o print do index - "BR", "RU", "IN", "CH", "SA"
    print(row)    #faz o print da informação de toda a panda Series

    
print("")
#se quisermos fazer o print de apenas a capital em cada iteração:
for label, row in BRICS.iterrows():
    print(label + "  " + row["capital"])

print("")
#Adicionar a coluna "name_lenght", com o nr de caracteres de cada país

for label, row in BRICS.iterrows():
    BRICS.loc[label, "name_lenght"] = len(row["country"])

print(BRICS)
print("")
#Dá, mas estamos a criar uma série a cada iteração

BRICS["name_lenght"] = BRICS["country"].apply(len)  # é como o lapply no R, ou sapply
print(BRICS)
```

#### Exercício 5 - Random walk

```python
#vídeo
#Random walk / Passeio aleatório

import numpy as np
np.random.seed(123)

outcome = []
for x in range(10):
    coin = np.random.randint(0,2)
    if coin == 0:
        outcome.append("heads")
    else:
        outcome.append("tails")
print(outcome)
print(outcome.count("tails")) #já agora
print("")
#Isto é um vetor aleatório mas não é um passeio aleatório, porque cada passo não depende do anterior
# Pode ser um passeio aleatório se registarmos o nr de "heads" que vai saindo:

import numpy as np
np.random.seed(123)

tails = [0]                       #No início não saiu nada, 0 "tails" (e 0 "heads" já agora)
for x in range(10):               #Para cada lançamento, até 10 lançamentos,
    coin = np.random.randint(0,2) # Para simplificar, admite-se que se sair 0 significa "heads" e 1 significa "tails"
    tails.append(tails[x] + coin) # Assim vai-se contando o nr de "tails" uma vez que se está a adicionar 1 pelo caminho
    
print(tails) # que significa o nr de "tails"

#Aqui o resultado que pretendíamos, o nr de "tails", dependia do nr (pseudo-)aleatório gerado anteriormente, por isso 
#trata-se de um passeio aleatório
```
